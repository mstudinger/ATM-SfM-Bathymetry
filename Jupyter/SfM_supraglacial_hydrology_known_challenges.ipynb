## 1) Caustics 
In optics, rippling caustics are visible features at the bottom of a body of shallow water that are commonly formed when light shines through waves on the surface of that body of water ([Caustics](https://en.wikipedia.org/wiki/Caustic_(optics)/)). This Jupyter notebook shows examples of caustics in natural-color images caused by surface waves on a supraglacial lake in Greenland.

```python
# load required Pythonâ„¢ modules and build file name for image
import os
from   PIL import Image
import matplotlib.pyplot as plt

# set input directory with image mosaic file
f_dir_imagery = r".." + os.sep + "data" + os.sep + "imagery" + os.sep + "challenges"
# set input file name for Jupyter notebook
f_name_rgb = f_dir_imagery + os.sep + "IOCAM0_2019_GR_NASA_20190506-131612.9216_caustics.jpg"
```

Caustics in natural-color images caused by surface waves on a supraglacial lake in Greenland. The lakes appear as sapphire-blue features to the eye and in natural color imagery. The caustics are the rippled features in the images at the bottom of the lake. 

```python
# import natural-color image mosaic and plot it
rgb_image = Image.open(f_name_rgb)

fig2, (ax1) = plt.subplots(1, 1, subplot_kw = {'xticks': [], 'yticks': []}) 
handle_ax1  = ax1.imshow(rgb_image)
```
